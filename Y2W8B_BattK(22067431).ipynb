```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#1.1 & 1.2 & 1.3
def f(y, t, k, m, b): # return derivatives of the array y

    return [ y[1], # the first equation dy[0]/dt
            -(k/m)*y[0] - (b/m)*y[1] # the second equation dy[1]/dt
           ]
k = 10.0
m = 1.0
b = 0.5
t = np.linspace(0, 10, 1000)
yinit = [0.1, 0] # TWO initial values, y and y'
y = odeint(f, yinit , t,args=(k, m, b))
plt.plot(t,y[:, 0], label='position')
plt.plot(t,y[:, 1], label='velocity')
plt.title('A mass on a spring velcoity/acceleration as a function of time')
plt.xlabel('time(s)')
plt.ylabel('velcoity (m/s)/Acceleration (m/s^2) ')
plt.legend()
```

```python
#2
#function
def f(xy, t, mu, A, omega):
    x, y = xy
    dxdt = y
    dydt = mu * (1 - x**2) * y - x + A * np.cos(omega * t)
    return [dxdt, dydt]

#parameters
mu = 1
A = 10
omega = 1

t_start = 0
t_end = 50
t = np.linspace(t_start, t_end, 5000)

yinit = [1.0, 0.0]

#solve of ODE
sol = odeint(f, yinit,t,args=(mu, A, omega))


plt.plot(sol[:, 0], sol[:, 1], lw=0.5)
plt.title("Phase portrait")
plt.xlabel("y")
plt.ylabel("y'")
plt.show()
plt.plot(t,sol[:, 1])
plt.title("y values verses t values")
plt.xlabel('t')
plt.ylabel('y')
```
